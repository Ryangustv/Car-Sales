Importando bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Carregando dados

```python
df = pd.read_csv("car_sales_data.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.title().str.strip().str.replace(" ", "_")
df.head()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
```

# Pair Plot


```python
sns.pairplot(numeric_df)
plt.suptitle("Pairplot de Variaveis Numericas", y=1.02)
plt.show()
```

# Correlação de Variáveis Númericas


```python
plt.figure(figsize=(10, 7))
correlation = numeric_df.corr()
sns.heatmap(correlation, annot=True)
plt.title("Correlações das Variáveis Númericas")
plt.show()
```

# Gráficos de Barras

```python
agrupado = df.groupby('Manufacturer')['Model'].size()
agrupado
plt.bar(agrupado.index, agrupado.values, color='skyblue', edgecolor='Black')
plt.title("Quantidade de Modelos por Marca")
plt.ylabel("Modelos")
plt.xlabel("Marca")
plt.show()
```

```python
fuel_type_counts = df.groupby('Fuel_Type').size()
plt.bar(fuel_type_counts.index, fuel_type_counts.values, color='skyblue', edgecolor='black')
plt.title("Distribuição de Tipos de Combustiveis")
plt.ylabel("Quantidade de Carros")
plt.xlabel("Tipos de Combustiveis")
plt.show()
```

# Histogramas

```python
plt.figure(figsize=(7,3))
sns.histplot(numeric_df['Engine_Size'], kde=True, bins=30)
plt.title("Histograma Engine Size")
plt.ylabel("Frequencia")
plt.xlabel("Engine size")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,3))
sns.histplot(numeric_df['Price'], kde=True, bins=40)
plt.title("Histograma de Preço")
plt.ylabel("Frequencia")
plt.xlabel("Preço")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,3))
sns.histplot(numeric_df['Year_Of_Manufacture'], kde=True, bins=20)
plt.title("Histograma de Ano de Fabricação")
plt.ylabel("Frequencia")
plt.xlabel("Ano de Fabricação")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,3))
sns.histplot(numeric_df['Mileage'], kde=True, bins=40)
plt.title("Histograma de kilometragem")
plt.ylabel("Frequencia")
plt.xlabel("Kilometragem")
plt.tight_layout()
plt.show()
```
